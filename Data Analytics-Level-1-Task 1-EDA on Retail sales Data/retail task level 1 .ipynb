```python
# Data cleaning
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
# Monthly Sales Trend
```

```python
df['Date']=pd.to_datetime(df['Date'])

df['Month']=df['Date'].dt.to_period('M')

monthly=df.groupby('Month')['Sales'].sum()

monthly.plot(figsize=(10,5))
plt.title("Monthly Sales")
plt.show()
```

```python
# Observation :
# Monthly sales show fluctuations throughout the year.
# Some months generated significantly higher revenue than others.
# The sales trend indicates seasonal variations in customer purchasing behavior.
```

```python
# Quarterly Sales Trend
```

```python
df['Quarter']=df['Date'].dt.to_period('Q')

quarter=df.groupby('Quarter')['Sales'].sum()

quarter.plot(figsize=(8,5))
plt.show()
```

```python
# Observation:
# Quarterly sales comparison shows differences in revenue across quarters.
# One quarter recorded the highest overall sales, indicating better business performance during that period.
# Quarterly analysis helps identify peak and low sales seasons.
```

```python
# Customer Demographics
```

```python
sns.countplot(x='Gender',data=df)
plt.show()
```

```python
# Observation:
# Both male and female customers contribute significantly to total sales.
# The customer distribution is fairly balanced between genders.
```

```python
sns.histplot(df['Age'],bins=10)
plt.show()
```

```python
# Observations:
# Adults represent the largest customer segment.
# Younger and older age groups contribute comparatively fewer purchases.
# Businesses can focus more on the dominant age group for targeted marketing.
```

```python
# Product Category vs Total Revenue
```

```python
category_sales = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title("Revenue by Product Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# Product Category Purchase Count
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Product Category')
plt.xticks(rotation=45)
plt.title("Number of Purchases by Product Category")
plt.show()
```

```python
# Observations:
# Product categories generate different levels of revenue.
# The highest-selling category contributes the largest share of total sales.
# Low-performing categories may require promotional offers or discounts.
```

```python
# Age vs Total Amount (Scatter Plot)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Age', y='Total Amount', hue='Gender')
plt.title("Age vs Total Amount")
plt.show()
```

```python
# Observation:
# Customers of different age groups spend different amounts.
# Spending patterns vary, but there is no strong linear relationship between age and total purchase amount.
```

```python
# Age Group vs Average Spending
```

```python
avg_spend = df.groupby('Age Group')['Total Amount'].mean()

plt.figure(figsize=(8,5))
avg_spend.plot(kind='bar')
plt.title("Average Spending by Age Group")
plt.ylabel("Average Total Amount")
plt.show()
```

```python
# Gender vs Total Revenue
```

```python
gender_sales = df.groupby('Gender')['Total Amount'].sum()

plt.figure(figsize=(6,4))
gender_sales.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel("")
plt.title("Revenue Contribution by Gender")
plt.show()
```

```python
# Observation:  female customers revenue contribution is High
```

```python
# Quantity Distribution
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Quantity'], bins=10, kde=True)
plt.title("Quantity Distribution")
plt.show()
```

```python
# Observation:
# Most customers purchase a small number of products per transaction.
# High-quantity purchases are relatively rare.
```

```python
# Price per Unit Distribution
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Price per Unit'], bins=15, kde=True)
plt.title("Price per Unit Distribution")
plt.show()
```

```python
# Observation:
# Product prices vary across different categories.
# Most products fall within a moderate price range.
```

```python
# Boxplot for Outliers
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(y=df['Total Amount'])
plt.title("Outliers in Total Amount")
plt.show()
```

```python
# Monthly Revenue
```

```python
monthly = df.groupby('Month')['Total Amount'].sum()

plt.figure(figsize=(10,5))
monthly.plot(marker='o')
plt.title("Monthly Revenue")
plt.show()
```

```python
# Quarterly Revenue
```

```python
quarterly = df.groupby('Quarter')['Total Amount'].sum()

plt.figure(figsize=(6,4))
quarterly.plot(kind='bar')
plt.title("Quarterly Revenue")
plt.show()
```

```python
# Correlation Heatmap
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(df[['Age','Quantity','Price per Unit','Total Amount']].corr(),
            annot=True,
            cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Observations:
# Total Amount is positively correlated with Quantity and Price per Unit.
# Age has little or no significant correlation with purchase amount.
#  Quantity and Price per Unit are the major factors influencing sales revenue.
```
